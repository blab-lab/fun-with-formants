```python
import numpy as np
import scipy.io as sio

exclusions = np.array([137, 265, 289, 320, 325, 352, 421, 423, 486, 496, 535, 536, 592, 606, 610, 615, 718, 781, 792])-1
exclusions = np.concatenate((exclusions,  [ 73, 140, 211, 214, 270, 283, 474, 512, 527, 558, 615, 712, 795]))

m = sio.loadmat("dataVals.mat", struct_as_record=False)
dataVals = m['dataVals'][0]
nTrials = len(dataVals)

NO_SHIFT = 1
UP = 2
DOWN = 3

SHIFTS = {
  NO_SHIFT: np.array([0,           0]),
  UP:       np.array([-101.4639,  77.5903]),
  DOWN:     np.array([98.4974,  -111.9339])
}


chunkBoundaries = ((0,25), (25,50), (50,75), (75,100))
nSamples = chunkBoundaries[-1][-1]

targetFreqs = np.array([np.column_stack((dataVals[trial].f1[:nSamples], dataVals[trial].f2[:nSamples])) 
                 for trial in range(nTrials) 
                 if len(dataVals[trial].f1) >= nSamples 
                    and trial not in exclusions 
                    and dataVals[trial].cond==NO_SHIFT])

chunks = []
for b in chunkBoundaries:
    chunk = {}
    chunk['ear'] = []
    chunk['mouth'] = []
    chunk['target'] = targetFreqs.mean(axis=0)[b[0]:b[1]].mean(axis=0)
    for t in range(len(dataVals)):
        v = dataVals[t]
        condition = v.cond[0][0]
        if condition != 2 or t in exclusions or len(v.f1) < nSamples: continue
        f1 = v.f1[b[0]:b[1]].mean()
        f2 = v.f2[b[0]:b[1]].mean()
        produced = np.array([f1, f2])
        chunk['mouth'].append(produced)
        chunk['ear'].append(produced + SHIFTS[condition])
    chunks.append(chunk)
print len(chunks), "chunks"
```

```python
import numpy as np
import pymc as mc
npa = np.array

mouthFactor =  [
  mc.Uniform('mouthFactor%s'%i, 0, 1, value=0) 
  for i in range(len(chunks)-1)
]

gain =  mc.Uniform('gain', -1, 1) 
def gap(n):
    chunkMouth = chunks[n]['mouth']
    chunkEar   = chunks[n]['ear']
    target     = chunks[n+1]['target']
    def gapfunc(mouthFactor, gain):
        ret = []
        for v in range(len(chunkMouth)):
            err = 0
            err += (target - chunkMouth[v]) *    mouthFactor
            err += (target - chunkEar[v]  ) * (1-mouthFactor)
            ret.append(err*gain)
        return ret
    return gapfunc

gaps = [
    mc.Deterministic(eval = gap(i),
                  name = 'gap%s'%i,
                  parents = {
                    'gain': gain, 
                    'mouthFactor': mouthFactor[i]
                    },
                  doc='gap %s'%i)
    for i in range(len(chunks)-1)
]

def chunkPrediction(n, TARGET_TYPE):
    chunk = chunks[n-1][TARGET_TYPE]
    def predict(gap):
        return chunk
        #ret = []
        #for v in range(len(gap)):
            #ret.append(chunk[v] + gap[v])
        #return ret
    return predict


earOffset=mc.Uniform('earOffset', -1000,1000)
mouthOffset=mc.Uniform('mouthOffset', -1000,1000)

mouthNoise=mc.Exponential('mouthNoise', 1,1)
earNoise=mc.Exponential('earNoise', 1,1)

for i in range(1,len(chunks)):
    mc.Normal('chunk%sMouthObservation'%i, 
                mouthOffset + chunks[i-1]['mouth'],
                pow(mouthNoise,-0.5), 
                value=chunks[i]['mouth'],
                observed=True)
    
    mc.Normal('chunk%sEarObservation'%i, 
               earOffset +chunks[i-1]['ear'],
                pow(earNoise,-0.5), 
                value=chunks[i]['ear'],
                observed=True)
    
print len(chunks[0]['ear'])
```

```python
S = mc.MCMC([gain, mouthNoise, earNoise, earOffset, mouthOffset, mouthFactor])
S.sample(40000, burn=500)
```
