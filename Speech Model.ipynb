```python
import numpy as np
import pymc as mc
npa = np.array

chunks = [{
    'target': npa([6,6]),
    'ear':   [npa([5,6]), npa([5,6]), npa([5,6]), npa([5,6]), npa([5,6]) ],
    'mouth': [npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]) ]
  }, {
    'target': npa([6,6]),
    'ear':   [npa([3,6]), npa([3,6]), npa([3,6]), npa([3,6]), npa([3,6])],
    'mouth': [npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]) ]
  }, {
    'target': npa([6,6]),
    'ear':   [npa([2,6]), npa([2,6]), npa([2,6]), npa([2,6]), npa([2,6]) ],
    'mouth': [npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]), npa([6,6]) ]
  }, ]

mouthFactor =  [
  mc.Uniform('mouthFactor%s'%i, 0, 1, value=0) 
  for i in range(len(chunks)-1)
]


gain =  mc.Uniform('gain', 0, 1) 
    
def gap(n):
    chunkMouth = chunks[n]['mouth']
    chunkEar = chunks[n]['ear']
    target = chunks[n+1]['target']
    def gapfunc(mouthFactor, gain):
        ret = []
        for v in range(len(chunkMouth)):
            gap = 0
            gap += (target - chunkMouth[v])*mouthFactor
            gap += (target - chunkEar[v])*(1-mouthFactor)
            ret.append(gap*gain)
        return ret
    return gapfunc

gaps = [
    mc.Deterministic(eval = gap(i),
                  name = 'gap%s'%i,
                  parents = {'gain': gain, 'mouthFactor': mouthFactor[i]},
                  doc='gap %s'%i)
    for i in range(len(chunks)-1)
]

def chunkPrediction(n, TARGET_TYPE):
    chunk = chunks[n-1][TARGET_TYPE]
    def predict(gap):
        ret = []
        for v in range(len(gap)):
            ret.append(chunk[v] + gap[v])
        return ret
    return predict


noise=mc.Exponential('noise', 1,1)

for i in range(1,len(chunks)):
    
    mc.Normal('chunk%sEarObservation'%i, 
                mc.Deterministic(eval = chunkPrediction(i, 'ear'),
                      name = 'chunk%sEarPrediction'%i,
                      parents = {'gap': gaps[i-1]},
                      doc='chunk %s prediction'%i),
                noise, value=chunks[i]['ear'], observed=True)
    
    mc.Normal('chunk%sMouthObservation'%i, 
                mc.Deterministic(eval = chunkPrediction(i, 'mouth'),
                      name = 'chunk%sMouthPrediction'%i,
                      parents = {'gap': gaps[i-1]},
                      doc='chunk %s prediction'%i),
                noise, value=chunks[i]['mouth'], observed=True)

@mc.deterministic
def mouthChanges(early=mouthFactor[0], late=mouthFactor[1]):
    return late-early
```

```python
S = mc.MCMC([noise, mouthChanges, mouthFactor])
S.sample(20000, burn=1000)       
```

```python
from pylab import hist, show, figure

figure()
hist(S.trace('mouthFactor0')[:], 50, color='b', alpha=.4)
hist(S.trace('mouthFactor1')[:], 50, color='r', alpha=.4)

figure()

delta_samples=S.trace('mouthChanges')[:]

hist(delta_samples, 50, color='y', alpha=.4)

print "Probability mouth sensitivity INCREASES over time: %.3f" % \
    (delta_samples > 0).mean()

figure()

hist(S.trace('noise')[:], 50, color='g', alpha=.4)
#S.stats()
```
